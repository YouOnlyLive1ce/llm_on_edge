```python
import os
import json
import platform
import psutil
from pathlib import Path
from typing import Dict, Any, Optional, List
from datetime import datetime

class ModelInfoExtractor:
    def __init__(self, model_path: str):
        self.model_path = Path(model_path)
        self.info = {}
        self.config = None
        
    def extract_all(self) -> Dict[str, Any]:
        """Extract all available information."""
        # Initialize with system info
        self.info = {
            "model_family": None,
            "exact_hf_model_id": None,
            "commit_hash": None,
            "base_vs_instruct": "instruct",
            "local_path": str(self.model_path.absolute()),
            "pruning_method": None,
            "sparsity": None,
            "quantization_method": None,
            "quantization_details": None,
            "backend": None,
            "backend_version": None,
            "device": None,
            "transformers_version": None,
            "cpu": "ryzen 5 5600",
            "cpu_arch": platform.processor(),
            "ram_gb": round(psutil.virtual_memory().total / (1024**3), 2),
            "os": f"{platform.system()} {platform.release()}",
            "thread_count": os.cpu_count(),
            "context_length": None,
            "model_architecture": None,
            "vocab_size": None,
            "num_layers": None,
            "hidden_size": None,
            "num_attention_heads": None,
            "batch_size": 1,
            "max_tokens": 64,
            "temperature": 1,
            "repetition_penalty": 1.0,
            "stop_sequence_token": None,
            "seed": None
        }
        
        # Check if it's a GGUF file or safetensors folder
        if self.model_path.suffix == '.gguf':
            self._extract_gguf_info()
        else:
            self._load_config()
            self._extract_from_config()
            self._get_transformers_version()
        
        return self.info
    
    def _extract_gguf_info(self):
        """Extract information from GGUF file."""
        try:
            # Try to import llama-cpp-python
            from llama_cpp import Llama
            
            # Load model to extract metadata
            llm = Llama(
                model_path=str(self.model_path),
                n_ctx=0,  # Don't load for inference, just metadata
                verbose=False
            )
            
            # Extract metadata
            metadata = llm.metadata
            
            # Model architecture
            if 'general.architecture' in metadata:
                arch = metadata['general.architecture']
                self.info["model_architecture"] = arch
                
                # Map to model family
                if 'llama' in arch.lower():
                    self.info["model_family"] = "Llama"
                elif 'olmo' in arch.lower():
                    self.info["model_family"] = "OLMo"
                elif 'mistral' in arch.lower():
                    self.info["model_family"] = "Mistral"
                elif 'qwen' in arch.lower():
                    self.info["model_family"] = "Qwen"
                else:
                    self.info["model_family"] = arch
            
            # Extract model parameters
            if f"{arch}.context_length" in metadata:
                self.info["context_length"] = int(metadata[f"{arch}.context_length"])
            if f"{arch}.vocab_size" in metadata:
                self.info["vocab_size"] = int(metadata[f"{arch}.vocab_size"])
            if f"{arch}.layer_count" in metadata:
                self.info["num_layers"] = int(metadata[f"{arch}.layer_count"])
            if f"{arch}.embedding_length" in metadata:
                self.info["hidden_size"] = int(metadata[f"{arch}.embedding_length"])
            if f"{arch}.attention.head_count" in metadata:
                self.info["num_attention_heads"] = int(metadata[f"{arch}.attention.head_count"])
            
            # Get HF model ID if available
            if 'general.name' in metadata:
                name = metadata['general.name']
                if 'olmo' in name.lower():
                    self.info["exact_hf_model_id"] = "allenai/OLMo-2-0425-1B-Instruct"
                elif 'smol' in name.lower():
                    self.info["exact_hf_model_id"] = "HuggingFaceTB/SmolLM2-1.7B-Instruct"
            
            # Extract quantization info from GGUF filename or metadata
            gguf_filename = self.model_path.name
            quant_patterns = {
                'IQ4_XS': 'IQ4_XS (4-bit, improved quantization)',
                'IQ4_NL': 'IQ4_NL (4-bit, no list)',
                'Q4_K_M': 'Q4_K_M (4-bit, medium)',
                'Q4_K_S': 'Q4_K_S (4-bit, small)',
                'Q5_K_M': 'Q5_K_M (5-bit, medium)',
                'Q5_K_S': 'Q5_K_S (5-bit, small)',
                'Q8_0': 'Q8_0 (8-bit)',
                'Q2_K': 'Q2_K (2-bit)',
                'Q3_K': 'Q3_K (3-bit)',
                'IQ1_S': 'IQ1_S (1-bit)',
                'IQ2_S': 'IQ2_S (2-bit)',
                'IQ3_S': 'IQ3_S (3-bit)'
            }
            
            for pattern, desc in quant_patterns.items():
                if pattern in gguf_filename:
                    self.info["quantization_method"] = desc
                    self.info["quantization_details"] = {
                        "format": "GGUF",
                        "quantization_type": pattern,
                        "description": desc
                    }
                    break
            else:
                # Try to get from metadata
                if 'general.file_type' in metadata:
                    self.info["quantization_method"] = f"GGUF (type {metadata['general.file_type']})"
            
            # Get token information
            if f"{arch}.tokenizer.eos_token_id" in metadata:
                self.info["stop_sequence_token"] = int(metadata[f"{arch}.tokenizer.eos_token_id"])
            
            # Set backend info
            self.info["backend"] = "llama.cpp"
            self.info["backend_version"] = self._get_llama_cpp_version()
            self.info["device"] = "cpu"  # or "cuda" if GPU is used
            
            # Close the model to free memory
            del llm
            
        except ImportError:
            # If llama-cpp-python is not installed, try basic file inspection
            self.info["backend"] = "llama.cpp"
            self.info["backend_version"] = "unknown (llama-cpp-python not installed)"
            self.info["device"] = "cpu"
            
            # Try to extract basic info from file size and name
            file_size_gb = self.model_path.stat().st_size / (1024**3)
            
            # Guess quantization from filename
            gguf_filename = self.model_path.name
            if 'IQ4_XS' in gguf_filename:
                self.info["quantization_method"] = "IQ4_XS (4-bit, improved quantization)"
                self.info["quantization_details"] = {"format": "GGUF", "quantization_type": "IQ4_XS"}
            elif 'Q4_K' in gguf_filename:
                self.info["quantization_method"] = "Q4_K (4-bit, K-quant)"
            elif 'Q5_K' in gguf_filename:
                self.info["quantization_method"] = "Q5_K (5-bit, K-quant)"
            elif 'Q8' in gguf_filename:
                self.info["quantization_method"] = "Q8_0 (8-bit)"
            
            print(f"Note: Install llama-cpp-python for full GGUF metadata extraction: pip install llama-cpp-python")
            print(f"File size: {file_size_gb:.2f} GB")
        
        # Set default context length if not found
        if not self.info["context_length"]:
            self.info["context_length"] = 4096  # Default for many models
    
    def _get_llama_cpp_version(self):
        """Get llama-cpp-python version."""
        try:
            import llama_cpp
            return llama_cpp.__version__
        except (ImportError, AttributeError):
            return "unknown"
    
    def _load_config(self):
        """Load config.json from model path."""
        config_path = self.model_path / "config.json"
        if config_path.exists():
            with open(config_path, 'r') as f:
                self.config = json.load(f)
        else:
            print(f"Warning: config.json not found in {self.model_path}")
    
    def _extract_from_config(self):
        """Extract all available information from config.json."""
        if not self.config:
            return
        
        # Model architecture/family
        architectures = self.config.get("architectures", [])
        if architectures:
            arch = architectures[0]
            self.info["model_architecture"] = arch
            # Extract model family
            if "Olmo2" in arch:
                self.info["model_family"] = "OLMo-2"
                self.info["exact_hf_model_id"] = "allenai/OLMo-2-0425-1B-Instruct"
            elif "Llama" in arch:
                self.info["model_family"] = "Llama"
                self.info["exact_hf_model_id"] = "HuggingFaceTB/SmolLM2-1.7B-Instruct"
            else:
                self.info["model_family"] = arch.replace("ForCausalLM", "").strip()
        
        # Check for pruning method from path
        local_path = str(self.info['local_path']).lower()
        if 'wanda' in local_path:
            self.info['pruning_method'] = 'wanda'
        elif 'sparsegpt' in local_path:
            self.info['pruning_method'] = 'sparsegpt'
        
        # Check for sparsity from path
        if '0.2' in local_path:
            self.info['sparsity'] = '0.2'
        elif '0.5' in local_path:
            self.info['sparsity'] = '0.5'
        
        # Model parameters
        self.info["context_length"] = self.config.get("max_position_embeddings")
        self.info["vocab_size"] = self.config.get("vocab_size")
        self.info["num_layers"] = self.config.get("num_hidden_layers")
        self.info["hidden_size"] = self.config.get("hidden_size")
        self.info["num_attention_heads"] = self.config.get("num_attention_heads")
        self.info["stop_sequence_token"] = self.config.get("eos_token_id")
        self.info["transformers_version"] = self.config.get("transformers_version")
        
        # Quantization (GPTQ specific)
        quant_config = self.config.get("quantization_config")
        if quant_config:
            self.info["quantization_method"] = quant_config.get("quant_method", "unknown")
            bits = quant_config.get("bits")
            group_size = quant_config.get("group_size")
            if bits and group_size:
                self.info["quantization_method"] = f"GPTQ (W{bits}, g{group_size})"
                self.info["quantization_details"] = {
                    "bits": bits,
                    "group_size": group_size,
                    "sym": quant_config.get("sym"),
                    "desc_act": quant_config.get("desc_act")
                }
            elif bits:
                self.info["quantization_method"] = f"GPTQ (W{bits})"
    
    def _get_transformers_version(self):
        """Get transformers and GPTQ backend versions."""
        try:
            import transformers
            self.info["transformers_version"] = transformers.__version__
        except ImportError:
            self.info["transformers_version"] = "unknown"
        
        # Check if it's a GPTQ model
        if not self.model_path.suffix == '.gguf' and self.config and self.config.get("quantization_config"):
            self.info["backend"] = "gptqmodel"
            try:
                meta = self.config.get("quantization_config", {}).get("meta", {})
                quantizer = meta.get("quantizer", [None])[0] if meta.get("quantizer") else None
                self.info["backend_version"] = quantizer or "unknown"
                self.info["device"] = meta.get("pack_impl", "cpu") if meta else "cpu"
            except:
                self.info["backend_version"] = "unknown"
                self.info["device"] = "cpu"
    
    def set_inference_params(self, 
                            batch_size: int = 1, 
                            max_tokens: int = 64,
                            temperature: float = None,
                            repetition_penalty: float = 1.0,
                            seed: int = None):
        """Set inference-time parameters."""
        self.info["batch_size"] = batch_size
        self.info["max_tokens"] = max_tokens
        if temperature is not None:
            self.info["temperature"] = temperature
        self.info["repetition_penalty"] = repetition_penalty
        if seed is not None:
            self.info["seed"] = seed

    def save_to_json(self, output_path: str):
        """Save information to JSON file."""
        Path(output_path).parent.mkdir(parents=True, exist_ok=True)
        with open(f"{output_path}.json", 'w') as f:
            json.dump(self.info, f, indent=2)
        print(f"Saved model summary to: {output_path}.json")
    
    def print_report(self):
        for key, value in self.info.items():
            if value is not None:
                # Format key for display
                display_key = key.replace('_', ' ').title()
                print(f"{display_key:30}: {value}")

# Example usage
if __name__ == "__main__":
    folder_path=Path('../../out')
    model_paths = models = [str(path) for path in folder_path.rglob("*.gguf")] + [str(path.parent) for path in folder_path.rglob("quantize_config.json")]
    
    for model_path in model_paths:
        if os.path.exists(model_path):
            print(f"\nProcessing: {model_path}")
            extractor = ModelInfoExtractor(model_path)
            info = extractor.extract_all()
            extractor.print_report()
            extractor.save_to_json(f"../../model_summaries/{Path(model_path).name}")
        else:
            print(f"Path not found: {model_path}")
```
